```python
import pandas as pd
```

```python
# save to csv
df = pd.read_csv('metapath2vec_accuaricies')
```

```python
df
```

```python
df.set_index('epoch', inplace=True)
```

```python
# make confidence interval for ['wl64_run1', 'wl64_run2', 'wl64_run3', 'wl64_run4', 'wl64_run5']
df['wl64_mean'] = df[['wl64_run1', 'wl64_run2', 'wl64_run3', 'wl64_run4', 'wl64_run5']].mean(axis=1)
df['wl64_std'] = df[['wl64_run1', 'wl64_run2', 'wl64_run3', 'wl64_run4', 'wl64_run5']].std(axis=1)
df['wl64_ci'] = 1.96 * df['wl64_std'] / (5 ** 0.5)
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))

# Plot walk lengths
plt.plot(df.index, df['walk_length: 20 - accuracy'], label='Walk Length: 20', color='orange', linewidth=1.5)
plt.plot(df.index, df['walk_length: 40 - accuracy'], label='Walk Length: 40', color='blue', linewidth=1.5)
plt.plot(df.index, df['walk_length: 60 - accuracy'], label='Walk Length: 60', color='red', linewidth=1.5)
plt.plot(df.index, df['walk_length: 80 - accuracy'], label='Walk Length: 80', color='purple', linewidth=1.5)

# Plot mean and confidence interval for 64 walk lengths
plt.plot(df.index, df['wl64_mean'], label='Walk Length: 64 - mean', color='green', linewidth=1.5)
plt.fill_between(df.index, df['wl64_mean'] - df['wl64_ci'], df['wl64_mean'] + df['wl64_ci'], color='gray', alpha=0.3, label='Walk Length: 64 - CI')

plt.legend(loc='lower right', fontsize=28)
plt.xlabel('Epoch', fontsize=30)
plt.ylabel('Accuracy', fontsize=30)
plt.title('Accuracy per Epoch', fontsize=30)
plt.grid()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlim(0, 10)
plt.ylim(0, 0.35)

plt.show()
```
